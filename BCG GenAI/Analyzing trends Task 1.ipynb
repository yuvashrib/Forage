# Import necessary Libraries

```python
import pandas as pd
```

# Read your CSV File witht he extracted data

```python
df = pd.read_csv('Extracted Data.csv')
```

```python
df
```

# Analyzing trends with pandas

## Calculating Year-by-Year growth rates for Total Revenue

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
```

```python
df
```

## Replacing NaN values with 0

```python
df.fillna(0, inplace=True)
df
```

## Calculating Year-by-Year growth rates for Net Income

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

## Calculating Year-by-Year growth rates for Total Assets

```python
df['Assets Growth (%)'] = df.groupby(['Company'])['Total Assets'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

## Calculating Year-by-Year growth rates for Total Liabilities

```python
df['Liabilities Growth (%)'] = df.groupby(['Company'])['Total Liabilities'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

```python
df['Cash Flow from Operating Activities'] = pd.to_numeric(df['Cash Flow from Operating Activities'], errors='coerce')
```

## Calculating Year-by-Year growth rates for Total Cash Flow Growth From Operations

```python
df['Cash Flow Growth(%) from operations'] = df.groupby(['Company'])['Cash Flow from Operating Activities'].pct_change() * 100
df.fillna(0, inplace=True)
df
```

# Save the file

```python
df.to_csv('Financial Data.csv', index=False)
```

# Summarize your Data based on company-wise metrics

```python
summary_company_wise = df.groupby('Company').agg({
    'Total Revenue': 'mean',  # Average revenue over the years
    'Net Income': 'mean',     # Average net income over the years
    'Total Assets': 'mean',   # Average total assets
    'Total Liabilities': 'mean',  # Average total liabilities
    'Cash Flow from Operating Activities': 'mean',  # Average cash flow
    'Revenue Growth (%)': 'mean',  # Average revenue growth
    'Net Income Growth (%)': 'mean',  # Average net income growth
    'Assets Growth (%)': 'mean',  # Average assets growth
    'Liabilities Growth (%)': 'mean',  # Average liabilities growth
    'Cash Flow Growth(%) from operations': 'mean'  # Average cash flow growth
})
```

```python
summary_company_wise
```

- Apple's performance has been relatively stable, but it has seen a small decline in key metrics over the period. Revenue, net income, assets, and cash flow from operations have all seen marginal declines, indicating that while the company remains strong, its growth has slightly contracted.
- Microsoft has experienced a significant decline across all growth metrics, with revenue and net income shrinking the most. While its financial position is still strong overall, the decline in assets and cash flow from operations indicates a period of contraction.
- Tesla’s revenue, net income, and assets have declined sharply over the period, reflecting substantial contraction. However, despite these declines, Tesla has managed to grow its cash flow from operating activities, which is a positive signal for its operations efficiency.
- 2021 was a challenging year, with significant declines in revenue, net income, and other financial metrics. It reflects an overall contraction for companies in this dataset during that year.
- 2022 saw some stabilization compared to 2021, though growth rates were still negative for most metrics. Notably, cash flow from operating activities grew by 7.74%, indicating improved operational efficiency across companies.
- 2023 shows no significant changes in growth metrics, suggesting a period of stability after the fluctuations of previous years. Companies seem to have reached a point of equilibrium, with no growth or decline in revenue, income, assets, and liabilities.

### Overall 
- Apple and Microsoft show stability but slight declines across most financial metrics, indicating mature markets with slower growth.
- Tesla experienced the sharpest declines in revenue, net income, and assets, reflecting either market volatility or company-specific challenges. However, its positive growth in cash flow from operations is a positive sign of operational health.
- 2021 was the most challenging year across all companies, with steep declines in revenue and profitability.
